- Запитались в ChatGPT як вивести піраміде з символів
>Для створення піраміди з символів "!" можна використовувати цикли. Ось приклад коду на Python, який буде виводити піраміду, починаючи з вашого початкового рядка:

```python
spaces = 5  # Початкова кількість пробілів
exclamation_marks = 1  # Початкова кількість знаків оклику

# Внешній цикл для кожного рядка піраміди
for i in range(5):  # Змініть 5 на бажану висоту піраміди
    # Друкуємо пробіли
    print(" " * spaces, end="")
    
    # Друкуємо знаки оклику
    print("!" * exclamation_marks, end="")
    
    # Збільшуємо кількість знаків оклику і зменшуємо кількість пробілів для наступного рядка
    exclamation_marks += 2
    spaces -= 1
    
    # Перехід на новий рядок
    print()
```

- знайомимось з списками та кортеджі
    > `List` - впорядкована колекція змінних типів даних;
    
    > `Tuple` - впорядкована колекція незмінних даних;

```python
l = [1, 1.2, "текст", a]
ll = list(l)

print("Початкове значення списку:",l)

l.append("Це кінець") # Всталяємо в кінець
l.insert(1, "Це буде на індексі 1") # Вставляємо на позицію 1
l.pop(-2) # видаляємо 5

print("Після змін, список має вигляд:",l)

l.extend(ll)
print("Якщо зчепити два списки, буде:",l)
```

```python
t = (1, 1.2, "текст", a, 1)
tt = tuple(t)

print("Початкове значення кортеджу:",t)

print("Навіть використовуючи зчеплення, ми не модифікували наш tuple, а просто повернули нове значення:", t.__add__((1,)))

print("Кількість 1 буде:",t.count(1))
print("Знайдемо під яким індексом знаходиться число 1:",t.index(1))
```

- знайомство з Set
> `Set` - набір унікальних, невпорядкованих даних

```python
s = {1, 1, "1", '1', 2, 'Кінець'}
print("Значення в Set будуть:", s)
print("Витягнути елемент по індексу неможливо, однак ми можемо просто взяти якийсь елемент за допомогою:",s.pop())
s.add('1') # Додали вже існуючий елемент
print("Якщо хочемо працювати з елементами Set то його перетворюють в list:", list(s), list(s)[1])
```

- знайомство з словниками
> Словники це завжди пара `ключ: значення`, де ключ завжди посвинен бути унікальним (число, буква, стрічка)

```python
a = "Ключ"
d = {1: "Це значення при ключі 1", "1": 1, a: "Це значення для ключа в змінній а", "a": "Це значення при ключі а"}

print("Введений словник буде мати вигляд:\n", d)

print("Щоб доступитись до якось елемента, вказуємо його ключ: ", d[1],
      "\nЯкщо ми не знаємо які існують ключі, то можна використати: ", d.keys(),
      "\n\nМожна витягнути всі значення,",
      "перетворити в список і працювати з елементами як з простими даними.",
      "\n(візьмемо другий елемент з утвореного списку, якому відповідає ключ 'Ключ):", list(d.values())[2],
      "\n\nАбо можна було це зробити простіше і просто використати ключ а:", d["Ключ"])
```
