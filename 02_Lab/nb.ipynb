- Запитались в ChatGPT як вивести піраміде з символів
>Для створення піраміди з символів "!" можна використовувати цикли. Ось приклад коду на Python, який буде виводити піраміду, починаючи з вашого початкового рядка:

```python
spaces = 5  # Початкова кількість пробілів
exclamation_marks = 1  # Початкова кількість знаків оклику

# Внешній цикл для кожного рядка піраміди
for i in range(5):  # Змініть 5 на бажану висоту піраміди
    # Друкуємо пробіли
    print(" " * spaces, end="")
    
    # Друкуємо знаки оклику
    print("!" * exclamation_marks, end="")
    
    # Збільшуємо кількість знаків оклику і зменшуємо кількість пробілів для наступного рядка
    exclamation_marks += 2
    spaces -= 1
    
    # Перехід на новий рядок
    print()
```

- робота з циклами
> `while` - умовний цикл - ітерації будуть виконуватись поки умова є істииною (може бути безмежним циклом)

> `for` - із задою кількістю ітерацій (повторів)

```python
from random import randint
n = ["Ігор", "Данило", "Назар"]
h = ["Рибалку","Ігри","Аніме"]

j = 0
while j < 3:
    print(">>> {name} любить {hobbi}".format(name=n[j],hobbi=h[randint(0,2)]))
    j+=1

print("="*10, "Це вже не входитиь у цикл, цикл закінчився.\nЦикл While не дуже підходить для визначеної кількості ітерацій.")

# навіть для for можна йти по різному
for j in range(len(n)):
    print(">>> {name} любить {hobbi}".format(name=n[j],hobbi=h[randint(0,2)]))

print("="*10,"Це вже виконається поза циклом")

print(f"Демонструємо що буде виконувати функція range: {list(range(3))}")

# варіант 2 для for (це вже правильне використання for)
for j in n:
    print(">>> {name} любить {hobbi}".format(name=j,hobbi=h[randint(0,2)]))

# Спростимо код ще більше
print("="*10)
for j in n:
    print(f">>> {j} любить {h[randint(0,2)]}")
```

- Знайомство з обробниками винятків

```python
# код призведе до помилки бо ми хочемо викликати метод притаманний лише для інт змінних інших класів
ll = [int(2), int(100000), str("10"), float(5000), bytes(1)]

for a in ll:
    print(f"Змінна а має тип {type(a)}.")
    # вкористання блоку IF для обробки помилок є неправильним! 
    if isinstance(a, int): # type(a) == int Це теж саме
        print(f"Кількість байт у змінній a={a.bit_count()}. Вивід через IF.")
    else:
        print("Змінна а не є типу інт!")
    # для обробки помилок ми повинні використовувати конструкцію try
    try:
        print(f"Кількість байт у змінній a={a.bit_count()}. Вивід через TRY.")
    except AttributeError as e:
        print(f"Виникла помилка, деталі: {e}")
        try:
            print(f"Кількість байт у змінній a={int(a).bit_count()}. Вивід через TRY.")
        except ValueError as e:
            print(f"Виникла помилка, деталі: {e}\nТут неможливо порахувати кількість байт у змінній.")
```

> Випадково генерую числа де може попадатись число 0, на яке не можна ділити

> За допомогою конструкції try ми можемо опрацювати помилку ділення на 0 та просто задати іншу логіку виконання даної операції

```python
li = [randint(-5,5) for _ in range(30)]
print(li)
for i in li:
    try:
        d = 5/i  
    except ZeroDivisionError:
        print(f"!!! Було згенеровано число {i} на яке ділити не можна!!!")
        d = None
    finally:
        print(f"Будемо ділити 5 на число {i}, результат буде: {d}.")
```

- знайомство з Функціями та Лямбдами
> існують іменні та анонімні/безіменні функції, і лямбда є безіменною!

> Запитались у  ChatGPT пояснення що таке функції:

У мові програмування "функція" - це блок коду, який виконує певну задачу або виконує певні дії. Функції є однією з основних будівельних блоків програм та дозволяють вам організовувати код у великих програмах на більш менш логічні та повторно використовувані частини.

Основні аспекти функцій:

1. **Назва**: Функція має унікальне ім'я, за допомогою якого до неї можна звертатися.

2. **Параметри (аргументи)**: Функція може приймати вхідні дані, які називаються параметрами або аргументами. Ці параметри використовуються всередині функції для виконання певних операцій.

3. **Код**: Функція містить набір інструкцій та операцій, які виконуються при виклику функції. Цей код реалізує конкретний функціонал.

4. **Повернення значення**: Функція може повертати значення, як результат її виконання. Це значення може бути використане в інших частинах програми.

5. **Виклик функції**: Для виконання функції необхідно її викликати з іншої частини програми, передаючи необхідні аргументи, якщо це потрібно.

6. **Повторне використання**: Функції дозволяють вам згрупувати код і використовувати його в інших частинах програми або навіть в інших програмах, забезпечуючи повторне використання коду.

Функції спрощують розробку програм, оскільки дозволяють поділити велику задачу на менші, керовані та більш зрозумілі частини. Вони також полегшують зміну та підтримку коду, оскільки зміни в функціях можуть впливати тільки на їхню внутрішню роботу, не впливаючи на іншу частину програми.

```python
# після def буде назва функції за якою ми можемо її викликати 
def name_of_my_fun(arg:str):
    # тіло функції, все що буде на рівні відступу буде виконуватись у функції
    print(f"У функцію було передано значення: {arg}")
    return True if arg != None else False
    # Після цієї команди більше ніяки операції у функції виконуватись не будуть
    print("Це ніколи не виконається")

print(f"Виклик функції з аргументами: {name_of_my_fun(1)}")

# функція без аргументів
def no_args_fun():
    return True

print(f"Друга функція повернула значення: {no_args_fun()}")
```

```python
# Щодо виклику функції, він здійснюється за допомогою ()
a = no_args_fun

print(a)

def function_call_function(f):
    print("у цю функцію ми передали інакшу фінкцію і тепер викливаємо її тут:")
    f()
    print("Передана функція буде викликаною.")

def internal_no_args():
    print("Ця функція буде переданою кудись.")

function_call_function(internal_no_args)
```

- а тепер спробуємо зробити Лямбду

```python
# прототипом Лямбди буде така функція
def lambda_prototype(a:int, b:float):
    return f"Множимо два отримані числа: {a * b}"

a = lambda_prototype # присвоюємо НЕ викликану функціяю у змінну

print(f"Проста функція {a}, далі йде її виклик:", a(2, 2.5))

# перепишемо цю функцію через Лямбду
lb = lambda a, b: f"Множимо два отримані числа: {a * b}"
print(f"Ми створили Лямбду {lb}, далі йде її виклик:", lb(2,2.5))
```

```python
# приклад використання Лямбд
la = lambda i: [i+j if i > 10 else i-j for j in range(i)]

la(11)
```
